# View prepared training data samples

```python
import os
from pathlib import Path
import sys
import tensorflow as tf
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

base_path = os.path.abspath(os.getcwd())
os.chdir(Path(base_path).parent)
print(f'Current dir is now {os.getcwd()}')

sys.path.insert(0, "./")
from modules.settings import get_settings
settings = get_settings()
```

```python
train_positive_file_path = os.path.join(settings.training_data_dir, 'train_positive.npy')
train_negative_file_path = os.path.join(settings.training_data_dir, 'train_negative.npy')

positive_features = np.load(train_positive_file_path)
negative_features = np.load(train_negative_file_path)
```

```python
np.random.shuffle(positive_features)
fig, axes = plt.subplots(8, 2, figsize=(15, 12))
axes = axes.flatten()
for spectrogram, ax in zip(positive_features[:16], axes):
    rotated_img = ndimage.rotate(spectrogram, 90)
    ax.imshow(rotated_img)
    
plt.show()
```

```python
np.random.shuffle(negative_features)
fig, axes = plt.subplots(8, 2, figsize=(15, 12))
axes = axes.flatten()
for spectrogram, ax in zip(negative_features[:16], axes):
    rotated_img = ndimage.rotate(spectrogram, 90)
    ax.imshow(rotated_img)
    
plt.show()
```
